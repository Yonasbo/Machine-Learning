```python
df = pd.read_csv('titanic/train.csv')
```

```python
def custom_round(x, base=5):
    if isinstance(x, str):
        return 'No_Data'
    return int(base * round(float(x)/base))
print(sum(df['Fare'].isna()))
df_corr= df.corr()
df_corr
```

```python
df_2=df[[ 'Survived', 'Pclass', 'Sex', 'Age', 'SibSp',
       'Parch', 'Fare', 'Cabin', 'Embarked']]
df_2['Cabin'] = df_2['Cabin'].fillna('0').apply(lambda x: x[0])
df_2['Cabin'] =df_2['Cabin'].astype("category")
df_2['Sex'] =df_2['Sex'].astype("category")
df_2['Embarked'] =df_2['Embarked'].astype("category")
df_2['Age']=df_2['Age'].fillna('No_Data').apply(lambda x: custom_round(x, base=5)).astype("category")
df_2['Fare']=df_2['Fare'].apply(lambda x: custom_round(x, base=10)).astype("category")
df_2
```

```python
data=df_2.copy()
#data=data.loc[data['Sex']=='female']
X=data.loc[:, data.columns != 'Survived']
y=data['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
{'colsample_bytree': 0.5265089436079646,
 'gamma': 2.263602669154118,
 'max_depth': 17.0,
 'min_child_weight': 5.0,
 'reg_alpha': 45.0,
 'reg_lambda': 0.1860238040941059,
 'tree_method': 0}
# create model instance
bst = XGBClassifier(tree_method="approx",eta=0.1,n_estimators=30,max_depth=5,enable_categorical=True)
# fit model
bst.fit(X_train, y_train)
# make predictions
bst.score(X_test,y_test)
```

```python
space={'tree_method': hp.choice('tree_method',('approx','hist')),
        'max_depth': hp.quniform("max_depth", 3, 25, 1),
        'gamma': hp.uniform ('gamma', 1,9),
        'reg_alpha' : hp.quniform('reg_alpha', 10,180,1),
        'reg_lambda' : hp.uniform('reg_lambda', 0,1),
        'eta': hp.uniform('eta',0.01,0.2),
        'colsample_bytree' : hp.uniform('colsample_bytree', 0.1,1),
        'min_child_weight' : hp.quniform('min_child_weight', 0, 40, 1),
        'n_estimators': 250,
        'seed': 0
    }

df_2=df[[ 'Survived', 'Pclass', 'Sex', 'Age', 'SibSp',
       'Parch', 'Fare', 'Cabin', 'Embarked']]
df_2['Cabin'] = df_2['Cabin'].fillna(0).apply(lambda l: l if l==0 else 1)
df_2['Embarked'] =df_2['Embarked'].astype("category")
df_2['Age'].loc[df_2['Sex']=='female']=df_2['Age'].loc[df_2['Sex']=='female'].apply(lambda l: l if not np.isnan(l) else np.random.choice(df_2.loc[(df_2['Sex']=='female')&(~df_2['Age'].isna())]['Age']))
df_2['Age'].loc[df_2['Sex']=='male']=df_2['Age'].loc[df_2['Sex']=='male'].apply(lambda l: l if not np.isnan(l) else np.random.choice(df_2.loc[(df_2['Sex']=='male')&(~df_2['Age'].isna())]['Age']))
df_2['Sex'] =df_2['Sex'].apply(lambda x: 0 if x=='male' else 1)
#df_2['Age']=df_2['Age'].apply(lambda x: custom_round(x, base=10))
#df_2['Fare']=df_2['Fare'].apply(lambda x: x if x<=100 else 100)
df_2['Fare']=df_2['Fare'].apply(lambda x: custom_round(x, base=10))
#data=data.loc[data['Sex']=='female']
```

```python
data=df_2.loc[df_2['Fare']<=120]

X=data.loc[:, data.columns != 'Survived']
y=data['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)

def objective(space):
    clf=xgb.XGBClassifier(tree_method=space['tree_method'], eta=space['eta'],
                    n_estimators =space['n_estimators'], max_depth = int(space['max_depth']), gamma = space['gamma'],
                    reg_alpha = int(space['reg_alpha']),min_child_weight=int(space['min_child_weight']),
                    colsample_bytree=int(space['colsample_bytree']),eval_metric="auc",enable_categorical=True)
    
    evaluation = [( X_train, y_train), ( X_test, y_test)]
    
    clf.fit(X_train, y_train,
            eval_set=evaluation, 
            verbose=False)
    

    pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, pred>0.5)
    print ("SCORE:", accuracy)
    return {'loss': -accuracy, 'status': STATUS_OK }
```

```python
df_2.loc[df_2['Fare']>100]
```

```python
trials = Trials()

best_hyperparams = fmin(fn = objective,
                        space = space,
                        algo = tpe.suggest,
                        max_evals = 100,
                        trials = trials)
```
